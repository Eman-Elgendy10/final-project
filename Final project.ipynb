```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('supermarket_sales.csv')
```

```python
data['Date'] = data['Date'].str.replace('/','-')
```

```python
data['Date'].head(10)
```

```python
data.rename(columns={'Total':'Total sales'},inplace=True)
```

```python
data.rename(columns={'Payment':'Payment methode'},inplace=True)
```

```python
data.head(10)
```

```python
data['GOGS/Sales']=data['cogs']/data ['Total sales']
```

```python
data['GOGS/Sales']
```

```python
data.isnull().sum()
```

```python
data.duplicated()
```

```python
Sales_by_products=data.groupby ('Product line')['Total sales'].sum()
```

```python
Sales_by_products
```

```python
Sales_by_products.sort_values(ascending=False).plot(kind='bar', color='g')
plt.xticks(rotation =75)
plt.grid( axis ='y', color ='g' , alpha= 0.35, linestyle =':')
plt.title('Sales by products', fontsize= 15)
plt.xlabel('products')
plt.ylabel('Sales')
sns.despine()
```

```python
Sales_by_gender=data.groupby ('Gender')['Total sales'].sum()
```

```python
Sales_by_gender
```

```python
Sales_per_month =data.groupby ('Date')['Total sales'].sum()
```

```python
plt.pie(Sales_by_gender,startangle=90, labels =Sales_by_gender.index, colors =['#2D3255','#f6B17A'], autopct='%1i%%')
plt.legend (labels =Sales_by_gender.values, loc ='upper left')
plt.title('Sales by gender', fontsize= 15)
```

```python
Top_products_by_quantity =data.groupby ('Product line')['Quantity'].sum()
```

```python
Top_products_by_quantity
```

```python
Top_products_by_quantity.sort_values(ascending=False).plot(kind='bar', color='purple')
plt.xticks(rotation =75)
plt.title('products by quantity', fontsize= 15)
plt.xlabel('products')
plt.ylabel('Quantity')
sns.despine()
```

```python
Top_rated_products =data.groupby ('Product line')['Rating'].mean()
```

```python

Top_rated_products
```

```python
Top_rated_products.sort_values(ascending=True).plot(kind='barh', color='darkred')
plt.xticks(rotation =90)
plt.title('Top rated products', fontsize= 15)
plt.xlabel('rating')
plt.ylabel('products')
sns.despine()
```

```python
data['Date']=pd.to_datetime(data['Date'])
Sales_per_month = data.groupby('Date')['Total sales'].sum().reset_index()
plt.figure(figsize=(10,5))
plt.title('Total sales by date', fontsize= 15)
plt.fill_between(Sales_per_month['Date'],Sales_per_month['Total sales'],color='skyblue', alpha=0.4)
plt.plot(Sales_per_month['Date'], Sales_per_month['Total sales'], color='blue', alpha=0.6)
```

```python
sales_by_city =data.groupby ('City')['Total sales'].sum()
```

```python
sales_by_city
```

```python
plt.figure (figsize=(8,6))
plt .pie(sales_by_city,labels= sales_by_city.index,startangle=90,autopct='%1i%%')
plt.title ('sales by city')
```

```python
perffered_payment_method =data.groupby ('Payment methode')['Invoice ID'].count()
```

```python
perffered_payment_method
```

```python
perffered_payment_method.sort_values(ascending=False).plot(kind='bar', color='orange')
plt.xticks(rotation =45)
plt.title('perffered payment method', fontsize= 15)
plt.xlabel('Payment methode')
plt.ylabel('sales')
sns.despine()
```
